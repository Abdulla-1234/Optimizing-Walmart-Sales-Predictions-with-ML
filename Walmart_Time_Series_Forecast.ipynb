# **Data Normalization**

```python
num_col = ['Weekly_Sales','Size','Temperature','Fuel_Price','CPI','Unemployment','Total_MarkDown','max','min','mean','median','std']
```

```python
minmax_scale = MinMaxScaler(feature_range=(0, 1))
def normalization(df,col):
  for i in col:
    arr = df[i]
    arr = np.array(arr)
    df[i] = minmax_scale.fit_transform(arr.reshape(len(arr),1))
  return df
```

```python
data.head()
```

```python
data = normalization(data.copy(),num_col)
```

# **Data Splitted into Training, Validation, Test**

```python
X = data.drop(['Weekly_Sales'],axis=1)
Y = data.Weekly_Sales
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.20, random_state=50)
```

# **Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()  # Remove the "normalize=False" argument
lr.fit(X_train, y_train)
```

```python
lr_acc = lr.score(X_test,y_test)*100
print("Linear Regressor Accuracy - ",lr_acc)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn import metrics
print("MAE" , metrics.mean_absolute_error(y_test, y_pred))
print("MSE" , metrics.mean_squared_error(y_test, y_pred))
print("RMSE" , np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("R2" , metrics.explained_variance_score(y_test, y_pred))
```

```python
lr_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
lr_df.to_csv('/content/drive/MyDrive/Walmart-Sales-Time-Series-Forecasting-Using-Machine-Learning-master/Walmart-Sales-Time-Series-Forecasting-Using-Machine-Learning-master/predictions/lr_real_pred.csv')
lr_df
```

```python
plt.figure(figsize=(20,8))
plt.plot(lr.predict(X_test[:200]), label="prediction", linewidth=2.0,color='blue')
plt.plot(y_test[:200].values, label="real_values", linewidth=2.0,color='lightcoral')
plt.legend(loc="best")
plt.savefig('/content/drive/MyDrive/Walmart-Sales-Time-Series-Forecasting-Using-Machine-Learning-master/Walmart-Sales-Time-Series-Forecasting-Using-Machine-Learning-master/plots/lr_real_pred.png')
plt.show()
```
